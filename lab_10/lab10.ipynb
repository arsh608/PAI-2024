```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

df = pd.read_csv('heart.csv')
df
```

```python
df.hist(figsize=(20,20))
plt.show()
```

```python
plt.figure(figsize=(10,8))
sn.heatmap(df.corr(),annot=True, cmap="YlGnBu", fmt=".2f")
plt.show()
```

```python
plt.pie(df['sex'].value_counts(), labels=['Male', 'Female'], autopct='%1.1f%%', colors= ["cornflowerblue","green"])
plt.show()
```

```python
plt.figure(figsize=(5,3))
sn.distplot(df['chol'])
plt.show()
```

```python
no_heart_disease = df[df['target'] == 0]
sn.set_style("whitegrid")
plt.figure(figsize=(10, 6))
plt.hist(no_heart_disease['age'], bins=15, color='steelblue', width = 2)
plt.xlabel('Age Range')
plt.ylabel('Population Count')
plt.xticks(range(25, 78, 4))
plt.show()
#no_heart_disease = df[df['target'] == 0]
# x_ticks = list(range(25, 78, 4))
# plt.figure(figsize=(10, 6))
# sn.histplot(no_heart_disease['age'], bins=20, kde=False)
# plt.xlabel('Age Range')
# plt.ylabel('Population Count')
# plt.title('Age Distribution for people with No heart disease')
# plt.xticks(x_ticks)
# plt.show()
```
