```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('insurance.csv')

print("First few rows of the dataset:")
print(df.head())

print("\nBasic information about the dataset:")
print(df.info())
print("\nStatistical summary of the dataset:")
print(df.describe())

#Check for outliers
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.boxplot(data=df, x='bmi')
plt.title("BMI Outliers")
plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='charges')
plt.title("Charges Outliers")
plt.subplot(2, 2, 3)
sns.boxplot(data=df, x='age')
plt.title("Age Outliers")
plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='children')
plt.title("Children Outliers")
plt.tight_layout()
plt.show()

#Distribution of continuous variables
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(df['bmi'], kde=True)
plt.title("Distribution of BMI")
plt.subplot(1, 3, 2)
sns.histplot(df['age'], kde=True)
plt.title("Distribution of Age")
plt.subplot(1, 3, 3)
sns.histplot(df['charges'], kde=True)
plt.title("Distribution of Charges")
plt.tight_layout()
plt.show()

# Correlation heatmap
numerical_df = df.select_dtypes(include=np.number)
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()

#Distribution of charges across different categories
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.boxplot(x='smoker', y='charges', data=df)
plt.title("Charges by Smoker Status")
plt.subplot(2, 2, 2)
sns.boxplot(x='region', y='charges', data=df)
plt.title("Charges by Region")
plt.subplot(2, 2, 3)
sns.boxplot(x='sex', y='charges', data=df)
plt.title("Charges by Gender")
plt.subplot(2, 2, 4)
sns.boxplot(x='children', y='charges', data=df)
plt.title("Charges by Number of Children")
plt.tight_layout()
plt.show()

#correlations
plt.figure(figsize=(10, 6))
sns.scatterplot(x='bmi', y='charges', hue='smoker', style='sex', data=df)
plt.title("Charges vs BMI, grouped by Smoker Status and Gender")
plt.show()

#charges based on age and smoker status
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='charges', hue='smoker', style='sex', data=df)
plt.title("Charges vs Age, grouped by Smoker Status and Gender")
plt.show()
```
