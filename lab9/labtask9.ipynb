```python
import pandas as pd
import matplotlib.pyplot as plt

iris_df = pd.read_csv(r'Iris.csv')

print(iris_df.head())

plt.plot(iris_df['SepalLengthCm'], iris_df['PetalLengthCm'], linestyle='-')
plt.title('Petal Length vs. Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

iris_df = pd.read_csv(r'Iris.csv')

species_counts = iris_df['Species'].value_counts()

plt.figure(figsize=(8, 5))
species_counts.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title('Number of Records for Each Iris Species')
plt.xlabel('Species')
plt.ylabel('Number of Records')
plt.xticks(rotation=0)
#plt.grid(axis='y')
#plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

iris_df = pd.read_csv(r'Iris.csv')
print(iris_df.head())

colors = {'Iris-setosa': 'blue', 'Iris-versicolor': 'orange', 'Iris-virginica': 'green'}

plt.scatter(iris_df['Species'], iris_df['PetalWidthCm'],
            c=iris_df['Species'].map(colors),
            alpha=0.7, edgecolors='w', s=100)

plt.title('Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')

plt.grid()
plt.tight_layout()

plt.show()
```
