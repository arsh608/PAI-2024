```python
#@title TASK 1

import matplotlib.pyplot as plt

students = [f'Student {i+1}' for i in range(20)]
heights = [150, 160, 155, 165, 170, 175, 160, 162, 158, 167,
           172, 155, 159, 174, 163, 168, 177, 169, 161, 165]

plt.figure(figsize=(10, 5))
plt.bar(students, heights, color=plt.cm.Set3.colors[:20])
plt.xticks(rotation=90)
plt.title("Student Heights Bar Chart")
plt.xlabel("Students")
plt.ylabel("Height (cm)")
plt.show()

plt.figure(figsize=(6, 6))
plt.pie(heights, labels=students, autopct='%1.1f%%', colors=plt.cm.Set3.colors[:20])
plt.title("Distribution of Heights")
plt.show()
```

```python
#@title TASK 2
cities = ['Karachi', 'Lahore', 'Islamabad', 'Quetta', 'Peshawar']
populations = [15400000, 11000000, 1200000, 1000000, 2000000]

plt.figure(figsize=(8, 5))
plt.barh(cities, populations, color='skyblue')
plt.xlabel("Population")
plt.ylabel("Cities")
plt.title("Population of Cities")
plt.show()
```

```python
#@title TASK 3

flavors = ['Vanilla', 'Chocolate', 'Strawberry', 'Mango', 'Butterscotch']
sales = [120, 100, 90, 80, 70]

plt.figure(figsize=(6, 6))
plt.pie(sales, labels=flavors, autopct='%1.1f%%', startangle=140)
plt.title("Ice Cream Sales Distribution")
plt.show()
```

```python
#@title TASK 4

ages = [19, 22, 25, 27, 24, 30, 31, 28, 29, 32, 34, 36, 38, 21, 23, 26, 35, 33, 20, 37]
age_groups = ['18-25', '26-30', '31-35', '36+']
counts = [sum(18 <= age <= 25 for age in ages),
          sum(26 <= age <= 30 for age in ages),
          sum(31 <= age <= 35 for age in ages),
          sum(age > 35 for age in ages)]

plt.figure(figsize=(6, 6))
plt.pie(counts, labels=age_groups, autopct='%1.1f%%')
plt.title("Student Age Distribution")
plt.show()
```

```python
#@title TASK 5

genders = ['Male', 'Female']
counts = [12, 8]

plt.figure(figsize=(6, 6))
plt.pie(counts, labels=genders, autopct='%1.1f%%', colors=['blue', 'pink'])
plt.title("Gender Distribution")
plt.show()
```

```python
#@title TASK 6

math_marks = [80, 85, 78, 92, 88, 76, 81, 94, 89, 75]
science_marks = [75, 87, 80, 85, 82, 78, 88, 91, 84, 79]

plt.scatter(math_marks, science_marks, color='green', label='Math vs Science')
plt.xlabel("Math Marks")
plt.ylabel("Science Marks")
plt.title("Math vs Science Comparison")
plt.legend()
plt.show()
```

```python
#@title TASK 7

import numpy as np

years = np.arange(1900, 2000)

np.random.seed(42)
increments = np.random.uniform(0.0, 0.3, size=len(years))

sea_levels = np.cumsum(increments)

plt.figure(figsize=(10, 5))
plt.scatter(years, sea_levels, color='blue', marker='o', label='Sea Level Rise')
plt.plot(years, sea_levels, linestyle='--', color='gray')

plt.xlabel("Year")
plt.ylabel("Sea Level Rise (cm)")
plt.title("Sea Level Rise Over the Past 100 Years")
plt.legend()
plt.show()

```

```python
#@title TASK 8

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [15, 25, 35, 45, 55]

plt.plot(x, y1, 'o-', color='pink', label='Line 1')
plt.plot(x, y2, 'o-', color='gray', label='Line 2')

plt.xlabel("Amazing X-axis")
plt.ylabel("Incredible Y-axis")
plt.title("Two Lines on One Graph")
plt.legend(loc='lower right')

plt.show()
```
